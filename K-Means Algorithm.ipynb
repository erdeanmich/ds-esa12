# K-Means Algorithm Playground

```python
import pandas as pd 
from sklearn.cluster import KMeans
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
```

First we load the dataset.

```python
df = pd.read_csv("data/edlich-kmeans-A0.csv")
df.head(5)
```

## Find out the optimal k with the elbow theory

Try out a range from 1 - 10

```python
sse = {} 
for k in range(1,10):
    kmeans = KMeans(n_clusters=k, max_iter=500).fit(df)
    sse[k] = kmeans.inertia_
```

Plot the sum of squared errors (SSE) in relation to the used k.

```python
plt.figure()
plt.plot(list(sse.keys()), list(sse.values()))
plt.xlabel("K")
plt.ylabel("SSE")
plt.show()
```

If we look at the plot we see that we probably have *5* clusters in our dataset. 

## Print the classification vector

```python
kmeans = KMeans(n_clusters=5)
kmeans = kmeans.fit(df)
labels = kmeans.predict(df)
labels
```
